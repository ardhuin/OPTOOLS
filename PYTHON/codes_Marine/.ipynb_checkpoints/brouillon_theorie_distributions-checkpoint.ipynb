```python
def test_independence(fx,thfx,ddlbins=50):
    # return True if "independent variables"
    hist_cos,binscos = np.histogram(fx,bins=ddlbins)
    hist_sin,binssin = np.histogram(thfx,bins=ddlbins)
    hist_joint0, bins1,bins2 = np.histogram2d(fx,thfx,bins=[binscos,binssin])
    hist_joint = hist_joint0/np.sum(hist_joint0)
    hist_cos2D,hist_sin2D = np.meshgrid(hist_cos,hist_sin,indexing='ij')
    hist_joint_multiply0 = hist_cos2D*hist_sin2D
    hist_joint_multiply2 = hist_joint_multiply0*np.sum(hist_joint0)/np.sum(hist_joint_multiply0)
    hist_joint_multiply = hist_joint_multiply0/np.sum(hist_joint_multiply0)
    err = (hist_joint0-hist_joint_multiply2)**2/hist_joint_multiply2

    fig,axs=plt.subplots(2,2,figsize=(20,12))
    axs[0,0].hist(binscos[:-1],binscos,weights=hist_cos)
    axs[0,0].hist(binssin[:-1],binssin,weights=hist_sin,alpha=0.5)
    axs[0,0].legend(('fx','thfx'))
    axs[0,0].set_xlabel('bins')
    axs[0,0].set_ylabel('counts')
    axs[0,0].grid(True)
    
#     axs[0,1].grid(False)
    im= axs[0,1].pcolormesh(err)
    _=plt.colorbar(im,ax=axs[0,1])
    axs[0,1].set_title('local difference |expected - obs|')
    axs[0,1].set_xlabel('fx - axis')
    axs[0,1].set_ylabel('thfx - axis')
    
#     axs[1,0].grid(False)
    im= axs[1,0].pcolormesh(hist_joint0)
    _=plt.colorbar(im,ax=axs[1,0])
    axs[1,0].set_title('observed distribution')
    axs[1,0].set_xlabel('fx - axis')
    axs[1,0].set_ylabel('thfx - axis')
    
#     axs[1,1].grid(False)
    im= axs[1,1].pcolormesh(hist_joint_multiply0)
    _=plt.colorbar(im,ax=axs[1,1])
    axs[1,1].set_title('distribution if independent vars')
    axs[1,1].set_xlabel('fx - axis')
    axs[1,1].set_ylabel('thfx - axis')
    
    
    critical = chi2.ppf(0.95,(ddlbins-1)*(ddlbins-1))
    value_out = np.sum(err[np.isfinite(err)])
#     return chi2_contingency([hist_joint_multiply0,hist_joint0])
#     return chisquare(hist_joint0,hist_joint_multiply2,ddof=hist_joint0.size,axis=None)#-1-4900)
    return value_out,critical,value_out<critical

def test_is_distrib(fx,distrib,nbins,args_distrib="",verbose=False):
    hist_obs,binsobs = np.histogram(fx,bins=nbins)
    xfordistrib = 0.5*(binsobs[:-1]+binsobs[1:])
    if verbose:
        print("hist_distrib0 = "+distrib+".pdf(xfordistrib"+args_distrib+")")
        print('LOCALS = ',[var for var in locals().keys()])
        print('GLOBALS = ',[var for var in globals().keys()])
    exec('hist_distrib0 = '+distrib+'.pdf(xfordistrib'+args_distrib+')',None,locals())
    
    if verbose:
        print('LOCALS after = ',[var for var in locals().keys()])
        print(locals()['hist_distrib0'])
    hist_distrib00 = locals()['hist_distrib0']
    hist_distrib = hist_distrib00*np.sum(hist_obs)/np.sum(hist_distrib00)
    plt.figure()
    plt.hist(binsobs[:-1],binsobs,weights=hist_obs,label='observed')
    plt.hist(binsobs[:-1],binsobs,weights=hist_distrib,label='tested distrib',alpha=0.5)
    plt.legend()
    plt.grid(True)
    err = (hist_obs-hist_distrib)**2/hist_distrib
    
    critical = chi2.ppf(0.95,(nbins-1)*(nbins-1))
    value_out = np.sum(err[np.isfinite(err)])
    if value_out<critical:
        print('H0 is not rejected : sample FOLLOWS the distribution')
    else:
        print('H0 is REJECTED : sample does NOT follow the distribution')
    return value_out,critical,value_out<critical,xfordistrib
    
#         hist_joint_multiply0*np.sum(hist_joint0)/np.sum(hist_joint_multiply0),hist_joint0,axis=None)
```

```python
def prep_dataset_for_interpolation(ds):
    # change variables and dimensions name  
    # --- for ribbons change main dimension name ---------------------
    is2sided=0
    if "n_phi" not in ds.dims:
        ds = ds.swap_dims({"time0":"n_phi"})
    
    omega=np.sqrt(ds["k_vector"]*9.81)
    ds=ds.assign(wave_spectra_fth_hs=ds['wave_spectra_kth_hs']*(2*np.pi)/(0.5*(9.81/omega)))       
    # print('PHI ds : ',ds["phi_vector"].values)
    # --- duplicate the half spectrum to have an entire 2D spectrum
    ds0 = ds.copy(deep=True)
    if is2sided==0:
        ds0["wave_spectra_kth_hs"].values = np.zeros(ds0["wave_spectra_kth_hs"].shape)
    ds0["phi_vector"].values = (ds0["phi_vector"].compute().data+180.)%360
    # print('PHI ds0 : ',ds0["phi_vector"].values)
    # --- concatenate the two parts of the spectrum ----------------
    ds2 = xr.concat([ds,ds0],dim="n_phi",data_vars='minimal')
    if is2sided==0:
        ds2["wave_spectra_kth_hs"].values = ds2["wave_spectra_kth_hs"].compute().data
    else:
        ds2["wave_spectra_kth_hs"].values = ds2["wave_spectra_kth_hs"].compute().data/2
    ds2 = ds2.sortby('phi_vector')
    # --- duplicate the first and last angle value for easy interpolation -------------
    ds3 = xr.concat([ds2.isel(n_phi=-1),ds2,ds2.isel(n_phi=0)],dim="n_phi",data_vars='minimal')
    # --- change the first and last new values to have a 2pi revolution ---------------
    A = np.concatenate([[-360],np.zeros((ds2.dims['n_phi'])),[360]])
    factor = xr.DataArray(A, dims="n_phi")
    ds3['phi_vector'].values = ds3['phi_vector']+factor
    ds3 = ds3.set_coords('phi_vector')
    return ds3

def prep_interp_grid(dkmin=0.0003734846,kmax=0.21):
    # --- define k interpolation vector (positive part) ----
    kX00 = np.arange(0.5*(dkmin),kmax,dkmin)
    # --- duplicate to have both positive and negative parts --------------
    kX0_origin = np.concatenate((kX00,-np.flip(kX00[:])))
    kX0 = np.fft.fftshift(kX0_origin)
    kx = kX0
    ky = kX0[:-2] 
    # --- generate 2D grid ------------------------
    kX,kY = np.meshgrid(kx,ky , indexing='ij')
    # --- compute associated K, Phi(in deg) ---------
    kK = np.sqrt(kX**2+kY**2)
    kPhi = np.arctan2(kY,kX)*180/np.pi
    kPhi[kPhi<0]=kPhi[kPhi<0]+360
    
    kK2 = xr.DataArray(kK, coords=[("kx", kx), ("ky",ky)])
    # print(kK2)
    kPhi2 = xr.DataArray(kPhi, coords=[("kx", kx), ("ky",ky)])
    kKkPhi2s = xr.Dataset(
            {'kK': kK2,
             'kPhi': kPhi2}
        ).stack(flattened=["kx", "ky"])#.reset_index("flattened")
    return kKkPhi2s

def get_1D_spec_from_obs(ds):
    kKkPhi2s = prep_interp_grid()
    ds3 = prep_dataset_for_interpolation(ds)
    
    Ekxky_0, kx_0, ky_0 = spectrum_to_kxky(1,np.squeeze(ds3['wave_spectra_kth_hs'].compute().data),
                                 ds3['k_vector'].compute().data,ds3["phi_vector"].compute().data)
    # define the spectrum as a dataArray to apply the interp
    Speckxky = xr.DataArray(Ekxky_0, dims=("nk", "n_phi"), coords={"nk": ds3['k_vector'].astype(np.float64),
                                                                  "n_phi":ds3["phi_vector"]})
    Speckxky_0 = Speckxky.assign_coords({'kx':(("nk", "n_phi"),kx_0),'ky':(("nk", "n_phi"),ky_0)})
 
    # --- apply interpolation -----------------------------------
    B=Speckxky.interp(nk=kKkPhi2s.kK,n_phi=kKkPhi2s.kPhi)
    B.name='Ekxky_new'
    B0 = B.reset_coords(("nk","n_phi"))
    Ekxky_da = B0.Ekxky_new.unstack(dim='flattened')
    dkx = xr.DataArray(np.gradient(Ekxky_da.kx), dims="kx")
    dky = xr.DataArray(np.gradient(Ekxky_da.ky), dims="ky")
    Ekx_order = (Ekxky_da*dky).sum(dim='ky')
    return Ekx_order,dkx
```

```python
# => S1_r and S1_i are independent => Thus
# env2 = S1_r**2+S1_i**2 ~ sigma**2 * chi2(2) => env ~ sigma * chi(2)
# chi2(k) : mean = k
#         : var = 2*k

# chi(k) : mean = sqrt(2)*gamma((k+1)/2) / gamma(k/2)
#        : var = k - mean**2

# chi(2) : mean = sqrt(2) * gamma(3/2)/gamma(1) 
#               = sqrt(2)*(sqrt(pi)/2) / 1
#               = sqrt( pi /2 )
#        : var = 2 - sqrt( pi/2 )**2
#              = 2 - (pi/2) = (4 - pi)/2


# sigma**2 * chi2(2) : mean = 2 * sigma**2
#                    : var = 4 * sigma**4

# sigma * chi(2) : mean = sqrt( pi /2 ) * sigma
#                : var = (4-pi)/2 * sigma**2

####
# Hs = 4*sigma => sigma = Hs/4
# 
# Env2 = Env**2 : mean = 2 * (Hs/4)**2 = Hs**2/2**3
#               : var = 4 * (Hs/4)**4 = Hs**4 / 2**6
#
# Env : mean = sqrt(pi/2) * Hs / 4 = sqrt(pi) * Hs / 2**2.5
#     : var = (4 - pi) / 2 * (Hs/4)**2 = (4 - pi) * Hs**2/ 2**5

# mean(env2)/mean(env) =  Hs**2 / 2**3 / (sqrt(pi) * Hs /2**2.5 )
#                      = Hs**2 * 2**2.5 / (sqrt(pi) * Hs * 2**3)
#                      = Hs /sqrt(pi * 2)

# var(env2) / var(env) = Hs**4 / 2**6 / ( (4- pi)*Hs**2 / 2**5)
#                      = Hs**2 * 2**5 / (4-pi) * 2**6
#                      = Hs**2 / 2*(4-pi)
```
