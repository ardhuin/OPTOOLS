```python
def filter_big_change_from_ref(X,Xref,thresh):
    condition1= (np.isfinite(X))&(X!=0)&(np.isfinite(Xref))&(Xref!=0)
    ind = np.where(condition1)[0]
    
    X0 = np.inf*np.ones(X.shape)
    X0[ind]=np.exp(np.abs(np.log(X[ind]/Xref[ind])))
    
    Xlim=thresh*np.nanmedian(np.abs(X0[np.isfinite(X0)]))

    ind_suspect1 = np.where(np.abs(X0)>Xlim)[0]
    X1 = np.copy(X)
    X1[ind_suspect1]=np.nan
    
    return X1

def CFOSAT_smooth_and_filter_track(time_alti,swh_alti,ind,kernel_scatter=7,kernel_median=17,scatter_threshold=1):
    isorder1 = 0
    isplot=1
    if isplot:
        fig,ax=plt.subplots(figsize=(16,5))
#         fig,ax=plt.subplots(2,1,figsize=(16,9),sharex=True)
    if isorder1: 
        # - STEP 1 : median filtering (with kernel median) ------
        swh_alti1=scs.medfilt(swh_alti,kernel_median)
        # - STEP 2 : filter Hs original if big changes compared to the median filtered timeserie ------
        swh_alti20 = filter_big_change_from_ref(swh_alti,swh_alti1,1.2)
        # - STEP 3 : removed values that are set to nan, are locally turned to 0 for the calculus of the scattering
        swh_alti200=np.copy(swh_alti20)
        swh_alti200[np.isnan(swh_alti200)]=0
        # - STEP 4 : Get the scattering metric based on the Hs where the big changes compared to ref are set to 0 ------
        scat2= get_scattering_info(swh_alti200,kernel_scatter,2)
        scat0= get_scattering_info(swh_alti200,kernel_scatter,0)
        scat1= get_scattering_info(swh_alti200,kernel_scatter,1)
                
        scatsfilt = scat0*scat1*scat2
        # - STEP 5 : apply filtering by applying a threshold to the scatter metric on the original 
        #            time series where big changes are set to nans --------------------
        indscat=np.where(scatsfilt>=scatter_threshold)[0]
        
        swh_alti2=np.copy(swh_alti20)
        swh_alti2[indscat]=np.nan
        
        if isplot:
            ax[0].plot(time_alti,swh_alti,'ok')
            ax[0].plot(time_alti,swh_alti1,'sb')    
            ax[0].plot(time_alti,swh_alti20,'*r')
            ax[0].plot(time_alti,swh_alti2,'.c')
            ax[0].grid(True)
            ax[0].set_title(time_alti[0].strftime("%Y/%m/%d"))
            ax[0].legend(('Step 0','Step 1 : median filter', 'Step 2: remove big changes from median',
                          'Step 3 : remove point with high scattering metric'))
            ax[1].plot(time_alti,scat2,'ok',zorder=10)
            ax[1].plot(time_alti,scat0,'*b',zorder=50)
            ax[1].plot(time_alti,scat1,'.r',zorder=100)
            ax[1].plot(time_alti,scat*scat0*scat1,'.c',zorder=101)
            ax[1].plot(time_alti,np.nanmedian([scat2,scat0,scat1],axis=0),'+g',zorder=102)
            ax[1].grid(True)
            ax[1].set_yscale('log')
            ax[1].legend(('Metric : Max-min','Metric : std ','Metric : RMS from median','Multiply all',
                         'Median of 3 metrics'))    
    else:
        # - STEP 1 : Get the scattering metric based on the Hs original TS ------
        scat2= get_scattering_info(swh_alti,kernel_scatter,2)
        scat0= get_scattering_info(swh_alti,kernel_scatter,0)
        scat1= get_scattering_info(swh_alti,kernel_scatter,1)
                
        scatsfilt = scat0*scat1*scat2
        # - STEP 2 : Filter by applying a theshold to the scatter metric on the original Hs TS -----
        indscat=np.where(scatsfilt>=scatter_threshold)[0]
        
        swh_alti20=np.copy(swh_alti)
        swh_alti20[indscat]=np.nan
        # - STEP 3 : Create a copy where the removed values are set to 0 instead of nan ---------- 
        swh_alti200=np.copy(swh_alti20)
        swh_alti200[np.isnan(swh_alti200)]=0
        
        # - STEP 4 : median filtering of the copy with removed values = 0 -----------------
        swh_alti1=scs.medfilt(swh_alti200,kernel_median)
        # - Filter the already filtered data by comparing to the median filtered TS -----
        swh_alti2 = filter_big_change_from_ref(swh_alti200,swh_alti1,1.2)
        
        if isplot:
            ax.plot(time_alti,swh_alti,'ok')
            ax.plot(time_alti,swh_alti20,'sc')    
            ax.plot(time_alti,swh_alti1,'*b')
            ax.plot(time_alti,swh_alti2,'.r')
            ax.grid(True)
            ax.set_title(time_alti[0].strftime("%Y/%m/%d"))
#             ax[0].legend(('Step 0','Step 1 : remove point with high scattering metric', 'Step 2: Median filtering',
#                           'Step 3 : remove big changes from median '))
            ax.legend(('Step 0','Step 1', 'Step 2',
                          'Step 3'))
    
#             ax[0].plot(time_alti,swh_alti,'ok')
#             ax[0].plot(time_alti,swh_alti20,'sc')    
#             ax[0].plot(time_alti,swh_alti1,'*b')
#             ax[0].plot(time_alti,swh_alti2,'.r')
#             ax[0].grid(True)
#             ax[0].set_title(time_alti[0].strftime("%Y/%m/%d"))
# #             ax[0].legend(('Step 0','Step 1 : remove point with high scattering metric', 'Step 2: Median filtering',
# #                           'Step 3 : remove big changes from median '))
#             ax[0].legend(('Step 0','Step 1', 'Step 2',
#                           'Step 3'))
#             ax[1].plot(time_alti,scat2,'ok',zorder=10)
#             ax[1].plot(time_alti,scat0,'*b',zorder=50)
#             ax[1].plot(time_alti,scat1,'.r',zorder=100)
#             ax[1].plot(time_alti,scat2*scat0*scat1,'.c',zorder=101)
#             ax[1].plot(time_alti,np.nanmedian([scat2,scat0,scat1],axis=0),'+g',zorder=102)
#             ax[1].grid(True)
#             ax[1].set_yscale('log')
#             ax[1].legend(('Metric : Max-min','Metric : std ','Metric : RMS from median','Multiply all',
#                          'Median of 3 metrics'))   
            plt.tight_layout()
            plt.savefig('/home/mdecarlo/PROJETS/Tempetes_vs_satellite/cfosat/timeseries_'+str(ind)+'.png')
    return  swh_alti2

```

```python
def get_rolling_values(x,kernel_size):
    lenx=len(x)
    rank_kernel=kernel_size//2
    B=np.zeros(lenx+2*rank_kernel)
    B[rank_kernel:-rank_kernel]=x
    
    indA = np.atleast_2d(np.arange(rank_kernel)) + np.atleast_2d(np.arange(lenx)).T
    
    return B[indA]

def get_scattering_info(x,kernel_size,type_metric=0):
    y=get_rolling_values(x,kernel_size)
    med = np.nanmedian(y,axis=1)

    if type_metric==0: # std
        r = np.std(y,axis=1)
    elif type_metric==1: # RMS from median
        r=np.sum((y-np.tile((med),(kernel_size//2,1)).T)**2,axis=1)
    elif type_metric==2: # Max-min
        r=(np.amax(y,axis=1)-np.amin(y,axis=1))/np.nanmedian(y,axis=1)
    return r
```
