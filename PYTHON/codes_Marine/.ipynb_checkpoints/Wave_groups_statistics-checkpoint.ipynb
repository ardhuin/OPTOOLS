```python
def moment_r(E_sig,sig,r):
    dsig = np.gradient(sig)
    mr = np.nansum(sig**r*E_sig*dsig)
    return mr

def moment_mu_r(E_sig,sig,r,sig_bar):
    dsig = np.gradient(sig)
    mu_r = np.nansum((sig-sig_bar)**r*E_sig*dsig)
    return mu_r

def psi_tau_val(E_sig,sig,x):
    dsig = np.gradient(sig)
    psi_tau = np.nansum(np.cos(sig*x)*E_sig*dsig)
    return psi_tau

def calc_psi_tau(E_sig,sig,x):
    dsig = np.gradient(sig)
    psi_tau = np.zeros(len(x))
    for ix,xval in enumerate(x):
        psi_tau[ix] = np.nansum(np.cos(sig*xval)*E_sig*dsig)
    return psi_tau
```

```python
#### SPATIAL DOMAIN ##########################################
dx = 10
nx = 4048
dkx = 2*np.pi/(dx*nx)
kp = k_from_f(0.1)
X = dx*np.arange(0,nx)

sigk = 0.001
kx = dkx*np.arange(-nx//2+1,nx//2+1)
Hs = 12
E_k = (Hs/4)**2/np.sqrt(2*np.pi)*np.exp((-0.5*(kx-kp)**2/sigk**2))
# E_k[:nx//2]=np.flip(E_k[nx//2-1:-1])
E_k_roll = np.roll(E_k,-int(nx//2))
sumZ1=4*np.sqrt(sum(E_k.flatten()*dkx)) 
print('Hs for Gaussian : ',sumZ1)
rng = np.random.default_rng(0)
rg = rng.uniform(low=0.0, high=1.0, size=(nx))

zhats=np.roll(np.sqrt(2*E_k*dkx)*np.exp(1j*2*np.pi*rg),-int(nx//2-1))
kx_roll=np.roll(kx,-int(nx//2-1)) # checks that kx2D(1,1)=0 ... 

eta = np.real(np.fft.ifft(zhats,norm="forward"))
eta_complex = np.fft.ifft(zhats,norm="forward")
```

```python
###### TIME domain ###############################
# transform spectrum in k into spectrum in sig using dispersion relation
dsigdk = group_speed_from_k(kx)
sig0 = sig_from_k(kx)     
E_sig_bad_sampling= E_k/dsigdk
# => sampling is not linear in sigma

# generate linear sampling in sigma --- 2 sided spectrum 
sigmax=sig0[-1]
dsig = 2*sigmax/nx
sig2sided = dsig*np.arange(-nx//2+1,nx//2+1)
E_sig2sided =np.zeros(np.shape(sig2sided))
f=spi.interp1d(sig0,E_sig_bad_sampling)#,kind='nearest') #Â Y interpolated along x
E_sig2sided[nx//2:]=f(sig2sided[nx//2:])
# E_sig2sided[:nx//2]=f(np.abs(sig2sided[:nx//2]))
# E_sig2sided[:nx//2]=np.flip(E_sig2sided[nx//2-1:-1])
# get the positive frequency part of the spectrum
E_sig = E_sig2sided[nx//2:]
sig = sig2sided[nx//2:]

dt = (2*np.pi)/(nx*dsig)
T = dt*np.arange(0,nx)

zhats_t=np.roll(np.sqrt(2*E_sig2sided*dsig)*np.exp(1j*2*np.pi*rg),-int(nx//2-1))
sig2_roll=np.roll(sig2sided,-int(nx//2-1)) # checks that kx2D(1,1)=0 ... 
print(sig2_roll[0])
eta_t_complex = np.fft.ifft(zhats_t,norm="forward")
eta_t = np.real(eta_t_complex)
```

```python
def compute_TFA2(Spec_shifted):# with 0-padding and rolling
    C2 = np.convolve(Spec_shifted,Spec_shifted,'full')
    n = len(Spec_shifted)

    ref =  np.zeros((3*n,1))
    ref[n:2*n,0] = Spec_shifted
    ref = np.repeat(ref,2*n-1,axis=1)
    
    ref_flip = np.zeros((3*n))
    ref_flip[:] = np.flip(ref[:,0])
    
    delayed = np.zeros((3*n,2*n-1))
    for i in range(2*n-1):
        delayed[:,i] = np.roll(ref_flip,i-n+1)
    
    prod_within_ref_bound = np.zeros((n,2*n-1))
    prod_within_ref_bound = delayed[n:2*n,:] * ref[n:2*n,:] 
    
    # define values for the axis in terms of shift wf and freq w
    # in order to sum w only between wf and 0 (or 0 and wf, depending on the sign of wf)
    wf,w = np.meshgrid(np.arange(-n+1,n),np.arange(-n//2,n//2))

    prod_within_ref_bound[(wf>0) & ((wf<w) | (w<0))]=0
    prod_within_ref_bound[(wf<0) & ((wf>w) | (w>0))]=0
    
    integral_Cf0f = np.sum(prod_within_ref_bound,axis=0)
    integral_Cf0f[n-1]=0
    
    CX = 2*(C2 - integral_Cf0f)

    return CX
```

```python
Env = np.abs(eta_complex)**2

Spec_env = np.abs(np.fft.fft(Env-np.mean(Env),norm='forward'))**2/dkx
Spec_env_shifted = np.fft.fftshift(Spec_env)

# ---- Spectrum of S1 ------------
Spec_no_env = np.abs(np.fft.fft(eta,norm='forward'))**2/dkx

Spec_no_env_nosq = np.sqrt(np.abs(np.fft.fft(eta,norm='forward'))**2/dkx) # |FFT(S1)| not squared !!! 
freq=np.fft.fftfreq(nx, d=dx)
Spec_shifted = np.fft.fftshift(Spec_no_env)
Spec_shifted_nosq  = np.fft.fftshift(Spec_no_env_nosq)

freq_shifted = np.fft.fftshift(freq)

Z1p = np.zeros(nx)
Z1p[nx//2:]=Spec_shifted[nx//2:]
Z1n = np.zeros(nx)
Z1n[0:nx//2]=Spec_shifted[0:nx//2]

C0 = np.convolve(2*Z1p,2*Z1n,'full')
C = ((np.abs(C0[nx//2:nx//2+nx]))*dkx)**2

CX0 = compute_TFA2(Spec_shifted_nosq)
CX = np.abs((CX0[nx//2:nx//2+nx])*dkx)

plt.figure(figsize=(12,6))
_=plt.plot(freq_shifted,Spec_env_shifted,label='spectrum of "envelope"')
_=plt.plot(freq_shifted,Spec_shifted,'--',label='spectrum of surface')
_=plt.plot(freq_shifted,C,label='C = .... ')
_=plt.plot(freq_shifted,CX,'--',label='CX = .... ')
_=plt.legend()
_=plt.ylim(10**-12,10**6)
_=plt.yscale('log')
_=plt.xlim((0,0.01))
# _=plt.ylim(10**-12,10**6)
_=plt.grid(True)

plt.figure(figsize=(12,6))
_=plt.plot(freq_shifted,Spec_env_shifted/C,label='ratio env/C')
_=plt.legend()

_=plt.yscale('log')
# _=plt.xlim((0,0.0005))

_=plt.grid(True)
```

```python
# eq 2.3     
x = np.linspace(-200,200,3000)
psi_tau = calc_psi_tau(E_sig,sig,x) 

# eq 2.6
moments = np.zeros(10)
for ik in range(10):
    moments[ik]=moment_r(E_sig,sig,ik)
    
# eq 2.7 (comp m0 vs psi_0)
psi_0 = psi_tau_val(E_sig,sig,0)   

# eq 2.8 sig_bar
sig_bar = (moments[1]/moments[0])

# eq 2.9 moments about the mean mu
moments_mu = np.zeros(10)
for ik in range(10):
    moments_mu[ik]=moment_mu_r(E_sig,sig,ik,sig_bar)

# eq 2.10
display(Math(r"\mu_0 = {} =?= m_0 = {}".format(moments_mu[0], moments[0])))
display(Math(r"\mu_1 = {} =?= 0".format(moments_mu[1])))
display(Math(r"\mu_2 = {} =?= m_2 - (m_1^2/m_0) = {}".format(moments_mu[2],moments[2] -(moments[1]**2/moments[0]))))

# eq 2.11 - width param
nu_squared = moments_mu[2]/(moments_mu[0]*sig_bar**2)
```
