```python
def myconv(x, h,axis=-1):
    assert np.shape(x) == np.shape(h), 'Inputs to periodic convolution '\
                               'must be of the same period, i.e., shape.'

    X = np.fft.fft(x,axis=axis)
    H = np.fft.fft(h,axis=axis)

    return np.real(np.fft.ifft(np.multiply(X, H),axis=axis))

prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

twopi = 2*np.pi
```

```python
# ---- Make spectrum ----------------------
nx = 2048*4
Hs = 2
Z1, kX,sk = def_spectrumG_for_surface_1D(nx=nx,dx=10,T0=7,Hs=Hs,sk_k0=0.1,D=None,verbose=False)

S1_r,S1_i,X,dkx = surface_1D_from_Z1kx(Z1,kX,0,nx=None,dx=None,dkx=None)

env2 = S1_r**2 + S1_i**2
env = np.sqrt(env2)

iskmy = 1
iskmx = 1
Spec_surf0 = np.abs(np.fft.ifftshift(np.fft.fft(S1_r,norm="forward")))
Spec_surf = np.zeros(np.shape(Spec_surf0))
Spec_surf[:nx//2] = 0
Spec_surf[nx//2:] = np.flip(Spec_surf0[:nx//2])+Spec_surf0[nx//2:]

Spec_env2 = np.abs(np.fft.ifftshift(np.fft.fft(env2,norm="forward")))
Spec_env = np.abs(np.fft.ifftshift(np.fft.fft(env-np.mean(env),norm="forward")))
if iskmy:
    dkx1 = dkx*1000/(2*np.pi)
    Spec_env2_th = 8*2*np.fft.fftshift(myconv(0.5*Z1*2*np.pi/1000,0.5*np.flip(Z1*2*np.pi/1000)))*dkx1
else:
    dkx1 = dkx
    Spec_env2_th = 8*2*np.fft.fftshift(myconv(0.5*Z1,0.5*np.flip(Z1)))*dkx1
    
if iskmx:
    kX1 = kX*1000/(2*np.pi)
else:
    kX1 = kX

plt.figure(figsize=(10,5))
# plt.plot(kX1,Z1*2*np.pi/1000,'--b',lw=2,label ='original spectrum')

plt.plot(kX1,(np.abs(Spec_surf)**2)/(2*dkx1),'-',lw=2,color=colors[0],label='Surface Spectrum DSP($\zeta$)')
plt.plot(kX1,(np.abs(Spec_env)**2)/(2*dkx1),'*',color=colors[1],label='Envelope Spectrum DSP($\eta$)')
plt.plot(kX1,(np.abs(Spec_env2)**2)/(2*dkx1),'.',color=colors[3],label = '$Env^2$ Spectrum DSP($\eta^2$)')
# plt.plot(kX1,Spec_env2_th,'-k',label='Theory DSP($\eta^2$) = DSP($\zeta$)$\otimes$ DSP($\zeta$)')
print('Var(env2) = ',np.var(env2),', var(env2) from Spec = ',np.sum(np.abs(Spec_env2)**2/(2)),
     ', var(env2) from th = ',np.sum(Spec_env2_th*dkx1))
plt.yscale('log')
# plt.xscale('log')
# plt.xlim((-0.12,0.12))
plt.legend(loc="upper left")
if iskmx:
    plt.xlabel('$k_x$ [$km^{-1}$]')
    plt.xlim((-20,20))
else:
    plt.xlabel('$k_x$ [rad $\cdot m^{-1}$]')
    plt.xlim((-0.12,0.12))
if iskmy:
    plt.ylabel('DSP [$m^2 \cdot km $]')
    plt.ylim((1e-6,1e1))
else:
    plt.ylabel('DSP [$m^2 \cdot rad^{-1} \cdot m $]')
    plt.ylim((1e-3,1e3))

plt.title('Raw spectra')
plt.savefig('Raw_spectra_surface_env_7s_0.png')
plt.figure(figsize=(10,5))
plt.plot(kX1,(np.abs(Spec_surf)**2)/(2*dkx1),'-',lw=2,color=colors[0],label='DSP($\zeta$)')
# plt.plot(kX1,(np.abs(Spec_env)**2)/(2*dkx1),'*',color=colors[1],label='DSP($\eta$)')

plt.plot(kX1,(np.abs(Spec_env)**2)*Hs**2/(2*(4-np.pi)*2*dkx1),'*',color=colors[1],label='Normalized $DSP(\eta) \cdot H_s^2 / (2 \cdot(4-\pi))$')
plt.plot(kX1,(np.abs(Spec_env2)**2)/(2*dkx1),'.',color=colors[3],label = 'DSP($\eta^2$)')
plt.plot(kX1,Spec_env2_th,'-k',label='Theory DSP($\eta^2$) = DSP($\zeta$)$\otimes$ DSP($\zeta$)')
print('Var(env2) = ',np.var(env2),', var(env2) from Spec = ',np.sum(np.abs(Spec_env2)**2/(2)),
     ', var(env2) from th = ',np.sum(Spec_env2_th*dkx1))

plt.yscale('log')
plt.xscale('log')
# plt.xlim((-0.12,0.12))
plt.legend(loc="lower left")
if iskmx:
    plt.xlabel('$k_x$ [$km^{-1}$]')
    plt.xlim((-20,20))
else:
    plt.xlabel('$k_x$ [rad $\cdot m^{-1}$]')
    plt.xlim((-0.12,0.12))
if iskmy:
    plt.ylabel('DSP [$m^2 \cdot km $]')
    plt.ylim((1e-6,1e1))
else:
    plt.ylabel('DSP [$m^2 \cdot rad^{-1} \cdot m $]')
    plt.ylim((1e-3,1e3))
plt.title('Normalized spectra')
plt.savefig('Normalized_spectra_surface_env_7s.png')
```

```python
Hss=np.arange(1,16)
colorsjet = plt.cm.jet(np.linspace(0,1,len(Hss)))
cNorm = mcolors.Normalize(vmin=Hss[0],vmax=Hss[-1])
scalarMap = cmx.ScalarMappable(norm=cNorm, cmap='jet')

fig,axs=plt.subplots(2,1,figsize=(6,8))
LC = 4000
klim = twopi/LC
Envs = np.zeros((2501,len(Hss)))

for ih, Hs in enumerate(Hss):
    T0=10
    sk_k0 = 0.1
    specex,kvec2,c = def_spectrumG_for_surface_1D(nx=2048,Hs=Hs, T0 = T0,sk_k0=sk_k0)
    k = (twopi**2)/(9.81*T0**2)
    plato = (Hs**2/(sk_k0*k))*twopi*7.5*1e-6
#     plato = Hs**2*(4-np.pi)*1e-3/(32*twopi*(sk_k0*k)**2)
    plato_gauss = Hs**2*(4-np.pi)*1e-3*np.sqrt(twopi)/(32*np.sqrt(2)*(sk_k0*k))
    
#     aire = plato_gauss*sigma*sqrt(2*pi) = Hs**2*(4-np.pi)/32
    
    
    kvecnew = np.linspace(kvec2[0],kvec2[-1],20001)

    x = specex
    f = spi.interp1d(kvec2,x)
    xnew = f(kvecnew)
    dkvecnew = np.gradient(kvecnew)

    B2 = 8*2*np.fft.fftshift(myconv(0.5*xnew,0.5*np.flip(xnew)))*dkvecnew
    sumB2 = np.sum(B2*np.gradient(kvecnew))

    axs[0].plot(kvec2*1000/(twopi),x*twopi/1000,color=colorsjet[ih])
    axs[0].set_title('Wave spectra')
    axs[0].set_ylabel('PSD($\zeta$) [$m^2 * km$]')
    axs[0].set_xlabel('$k_x = 1/wavelength$ [$km^{-1}$]')
    axs[0].set_xlim((0,25))
    
    
    axs[1].plot(kvecnew*1000/(twopi),(8*(B2*2*(4-np.pi)/Hs**2))*twopi/1000,'-',color=colorsjet[ih])#,label='T = '+str(T1)+' s')

    axs[1].set_xscale('log')
#     axs[1].axhline(plato,color=colorsjet[ih],linestyle='--')
    axs[1].axhline(8*plato_gauss,color=colorsjet[ih],linestyle=':')
#     axs[1].axvline(1/4000,color='k')
    axs[1].axvline(1/1000,linestyle='--',color='k')
    axs[1].axvline(1/10000,linestyle='--',color='k')
    axs[1].axvline(1/100000,linestyle='--',color='k')
    axs[1].set_xlabel('1/wavelength [$km^{-1}$]')
    axs[1].set_ylabel('PSD($H_s$) [$m^2 * km$]')
    axs[1].set_title('PSD(Hs)')
#     Envs[:,ih]= ((B2*2*(4-np.pi)/Hs**2))*np.pi/500
    
    

plt.tight_layout()
cb = plt.colorbar(scalarMap,label='Hs [m]',ax=axs[:])
plt.savefig('PSD_HS_variation_withHS.png')
```
