```python
from functions_alti_simulator import *
import scipy.special as sps # function erf
import multiprocessing as mp
# import random
import warnings
from IPython.display import clear_output
from Misc_functions import extract_Welch_tiles_1D
from scipy import stats

warnings.filterwarnings("ignore")
mpl.rcParams.update({'figure.figsize':[10,6],'axes.grid' : True,'font.size': 14,'savefig.facecolor':'white'})
```

```python
# investigate variances r/r partial integration of spectrum #######################
nx = 2**14
dx=5
isplot=0
Hss = np.arange(4,20,4)
nHs=len(Hss)
T0=12
Lp=9.81/np.pi*T0**2;
kp=2*np.pi/Lp;
sx=kp*0.14
sy=sx;

nseed= 200

nLL = 20
L_low = 2*np.pi/(1.72*sx)#2.5*sx)
LLs=np.flip(np.linspace(500,8000,nLL))
Ntmax = int(2*(nx//(LLs.min()//dx))-1)
nhisto = 200

Hs_from_surf1D = np.zeros((nHs,nseed))
sx_Hs = np.zeros(nHs)
surf_histo = np.zeros((nHs,nseed,nhisto))
env_histo = np.zeros((nHs,nseed,nhisto))
bins_histo = np.zeros((nHs,nhisto+1))
random_surf_var_vec = np.zeros((nLL,nHs,nseed,Ntmax))
random_surf_std_vec = np.zeros((nLL,nHs,nseed,Ntmax))
random_surf_mean_vec = np.zeros((nLL,nHs,nseed,Ntmax))

footprint_surf_var_vec = np.zeros((nLL,nHs,nseed,Ntmax))
footprint_surf_var_0_vec = np.zeros((nLL,nHs,nseed,Ntmax))
footprint_surf_std_vec = np.zeros((nLL,nHs,nseed,Ntmax))
footprint_surf_mean_vec = np.zeros((nLL,nHs,nseed,Ntmax))
footprint_env_var_vec = np.zeros((nLL,nHs,nseed,Ntmax))
footprint_env_std_vec = np.zeros((nLL,nHs,nseed,Ntmax))
footprint_env_mean_vec = np.zeros((nLL,nHs,nseed,Ntmax))
Ntiles_all = np.zeros(nLL)
spec_funcs = []

for ihs,Hs in enumerate(Hss):
    Z1, kX, sx = def_spectrumG_for_surface_1D(nx=nx,dx=dx,T0=T0,Hs=Hs,sk_k0=0.1,D='None')
    sx_Hs[ihs] = sx
#     spec_funcs0 = []
    for isee in range(nseed):
        # Makes a surface with random phases using a seed (4th argument)
        S1_r,S1_i,X,dkx = surface_1D_from_Z1kx(Z1,kX,isee,nx=None,dx=None,dkx=None)
        
        rng = np.random.default_rng(isee)
        S_random = rng.normal(0,Hs/4,(nx))
        ### Envelope squared A2D and envelope B2D ###############################################
        A1D = (S1_r**2+S1_i**2)
        B1D = np.sqrt(A1D)
        A1D = A1D-np.mean(A1D)
        
        
        Hs_from_surf1D[ihs,isee] = 4*np.std(S1_r)
        if isee==0:
            bins0 = np.linspace(-2*Hs,2*Hs,nhisto+1)
            counts,bins=np.histogram(S1_r,bins=bins0)
            counts2,_=np.histogram(B1D,bins=bins0)

            print(bins.shape)
            surf_histo[ihs,isee,:]=counts
            env_histo[ihs,isee,:]=counts2
            bins_histo[ihs,:]=bins
        else:
            surf_histo[ihs,isee,:],bins=np.histogram(S1_r,bins=bins0)
            env_histo[ihs,isee,:],_=np.histogram(B1D,bins=bins0)
        # get histo of elevation !
        Hs_from_surf1D[ihs,isee] = 4*np.std(S1_r)

        for il,LL in enumerate(LLs):
            Nfoot = int(LL//dx) #Â size of footprint
            kc=2*np.pi/(LL)

            footprints_S1D, Ntiles = extract_Welch_tiles_1D(S1_r,Nfoot)
            footprints_Env, Ntiles = extract_Welch_tiles_1D(B1D-B1D.mean(),Nfoot)
            footprints_random, Ntiles = extract_Welch_tiles_1D(S_random,Nfoot)
            
            Ntiles_all[il] = Ntiles
            footprint_surf_var_vec[il,ihs,isee,:Ntiles] = np.var(footprints_S1D,axis=1,ddof=1)
            footprint_surf_var_0_vec[il,ihs,isee,:Ntiles] = np.var(footprints_S1D,axis=1,ddof=0)
            footprint_surf_std_vec[il,ihs,isee,:Ntiles] = np.std(footprints_S1D,axis=1)
            footprint_surf_mean_vec[il,ihs,isee,:Ntiles] = np.mean(footprints_S1D,axis=1)

            footprint_env_var_vec[il,ihs,isee,:Ntiles] = np.var(footprints_Env,axis=1,ddof=1)
            footprint_env_std_vec[il,ihs,isee,:Ntiles] = np.std(footprints_Env,axis=1)
            footprint_env_mean_vec[il,ihs,isee,:Ntiles] = np.mean(footprints_Env,axis=1)
            
            random_surf_var_vec[il,ihs,isee,:Ntiles] = np.var(footprints_random,axis=1,ddof=1)
            random_surf_std_vec[il,ihs,isee,:Ntiles] = np.std(footprints_random,axis=1)
            random_surf_mean_vec[il,ihs,isee,:Ntiles] = np.mean(footprints_random,axis=1)

    
#      spec_funcs.append(spec_funcs0)
```

```python
# ---- plot histo var of surf
var_chi2_surf = np.zeros((nHs,nLL,nseed))
mean_chi2_surf = np.zeros((nHs,nLL,nseed))
ddl_chi2_surf = np.zeros((nHs,nLL,nseed))
sigma_chi2_surf = np.zeros((nHs,nLL,nseed))

# sigma_norm_surf = np.zeros((nHs,nLL,nseed))

var_chi2_env = np.zeros((nHs,nLL,nseed))
mean_chi2_env = np.zeros((nHs,nLL,nseed))
ddl_chi2_env = np.zeros((nHs,nLL,nseed))
sigma_chi2_env = np.zeros((nHs,nLL,nseed))
sigma_rayl_env = np.zeros((nHs,nLL,nseed))
sigma_gamma_env = np.zeros((nHs,nLL,nseed))

# 
# env ~ Rayleigh(sig) => var env ~ Gamma(N,2sig**2)
# Rayleigh: estimateur sig (= Hs/4) = sqrt( (1/(2N)) * sum_i X_i**2)
# Gamma(k,theta) : esp = k*theta, var = k*theta**2
# var/esp = (k*theta**2)/(k*theta) = theta = 2*sig**2
# sig = sqrt(theta/2) = sqrt(0.5*var/esp)

var_chi2_rand = np.zeros((nHs,nLL,nseed))
mean_chi2_rand = np.zeros((nHs,nLL,nseed))
ddl_chi2_rand = np.zeros((nHs,nLL,nseed))
sigma_chi2_rand = np.zeros((nHs,nLL,nseed))
# sigma_norm_rand = np.zeros((nHs,nLL,nseed))

fig,axs=plt.subplots(2,2)
fig2,axs2=plt.subplots(2,2)


for ih,Hs in enumerate(Hss):
    for ill,LL in enumerate(LLs):
        for isee in range(nseed):
            A = np.squeeze(footprint_surf_var_vec[ill,ih,isee,:int(Ntiles_all[ill])])
            var_chi2A = np.var(A)
            mean_chi2A = np.mean(A)
            ddlA = 2*(mean_chi2A**2/var_chi2A)#/Ntiles_all[ill]
            
            var_chi2_surf[ih,ill,isee] = var_chi2A
            mean_chi2_surf[ih,ill,isee] = mean_chi2A
            ddl_chi2_surf[ih,ill,isee] = ddlA
            sigma_chi2_surf[ih,ill,isee] = (var_chi2A*ddlA/2)**(1/4)#*(ddl0)/(ddl0-1)
            
            B = np.squeeze(footprint_env_var_vec[ill,ih,isee,:int(Ntiles_all[ill])])
            B1 = np.squeeze(footprint_env_mean_vec[ill,ih,isee,:int(Ntiles_all[ill])])
            
            var_chi2B = np.var(B)
            mean_chi2B = np.mean(B)
            ddlB = 2*(mean_chi2B**2/var_chi2B)#/Ntiles_all[ill]
            
            var_chi2_env[ih,ill,isee] = var_chi2B
            mean_chi2_env[ih,ill,isee] = mean_chi2B
            ddl_chi2_env[ih,ill,isee] = ddlB
            sigma_chi2_env[ih,ill,isee] = (var_chi2B*ddlB/2)**(1/4)#*(ddl0)/(ddl0-1)
            sigma_rayl_env[ih,ill,isee] = np.mean(np.sqrt(0.5*(B+B1**2)))
            sigma_gamma_env[ih,ill,isee] = np.sqrt(0.5*var_chi2B/mean_chi2B)
            C = np.squeeze(random_surf_var_vec[ill,ih,isee,:int(Ntiles_all[ill])])
            
            var_chi2C = np.var(C)
            mean_chi2C = np.mean(C)
            ddlC = 2*(mean_chi2C**2/var_chi2C)#/Ntiles_all[ill]
            
            var_chi2_rand[ih,ill,isee] = var_chi2C
            mean_chi2_rand[ih,ill,isee] = mean_chi2C
            ddl_chi2_rand[ih,ill,isee] = ddlC
            sigma_chi2_rand[ih,ill,isee] = (var_chi2C*ddlC/2)**(1/4)#*(ddl0)/(ddl0-1)

            
            if isee == 0:
                nA,bins,_=axs[ih//2,ih%2].hist(A,density=True,histtype='step',
                                              color=plt.cm.jet(LL/LLs[0]),alpha=0.5,zorder=0)
                nB,binsB,_=axs[ih//2,ih%2].hist(B,density=True,histtype='step',
                                              color=plt.cm.jet(LL/LLs[0]),alpha=0.5,zorder=0)
                nC,binsC,_=axs[ih//2,ih%2].hist(C,density=True,histtype='step',
                                              color=plt.cm.jet(LL/LLs[0]),alpha=0.5,zorder=0)
                
                
            else:
                nA0,bins,_=axs[ih//2,ih%2].hist(A,bins=bins,density=True,histtype='step',
                                              color=plt.cm.jet(LL/LLs[0]),alpha=0.5,zorder=1)
                nB0,binsB,_=axs[ih//2,ih%2].hist(B,bins=binsB,density=True,histtype='step',
                                              color=plt.cm.jet(LL/LLs[0]),alpha=0.5,zorder=1)
                
                nC0,binsC,_=axs[ih//2,ih%2].hist(C,bins=binsC,density=True,histtype='step',
                                              color=plt.cm.jet(LL/LLs[0]),alpha=0.5,zorder=1)
                
                
                nA = nA + nA0
                nB = nB + nB0
                nC = nC + nC0
                               
            
        _=axs2[ih//2,ih%2].hist(bins[:-1],bins,weights=nA/nseed,histtype='step',
                                              color=plt.cm.jet(LL/LLs[0]),linewidth=2,zorder=10)
        _=axs2[ih//2,ih%2].hist(binsC[:-1],binsC,weights=nC/nseed,histtype='step',linestyle='--',
                                              color=plt.cm.jet(LL/LLs[0]),linewidth=2,zorder=10)
        
    _=axs2[ih//2,ih%2].set_yscale('log')
    axs[ih//2,ih%2].set_title('Hs = '+str(Hs))    
```

```python
#### 1D  Surface(t) #########################################################

####### Set-up the problem ####################################""
dt = 1
n = 2048
Tvec=dt*np.arange(0,n)
dk = 2*np.pi/(dt*n)
sig = 0.02

f = np.fft.fftfreq(n, d=dt)
freq_shifted = np.fft.fftshift(f)
Tp = 16
fp = 1/Tp
Hs = 18
##Â Define a spectrum #################################
E_k = (Hs/4)**2/np.sqrt(sig*2*np.pi)*np.exp((-0.5*(f-fp)**2/sig**2))

sumZ1=4*np.sqrt(sum(E_k.flatten()*dk)) 
print('Hs = ',sumZ1)

### Generate surface(t) ##################################
rng = np.random.default_rng(1)
rg = rng.uniform(low=0.0, high=1.0, size=(n))

zhats=(np.sqrt(2*E_k*dk)*np.exp(1j*2*np.pi*rg))
eta_c = np.fft.ifft(zhats,norm="forward")
eta_r = np.real(eta_c)
print('Hs surface = ',4*np.std(eta_r))
eta_i = np.imag(eta_c)
### Compute enveloppe ##################################################################
A = (eta_r**2+eta_i**2) # eta_r is real part, eta_i is imaginary part
B = A - np.mean(A) # B : surface enveloppe (squared)


### Go to frequency domain (DFT = Discrete Fourier Transform) ##########################
# with the conventions that the normalization is made 'forward',
# i.e. DFT(x)(k)=X(k) = 1/N * sum_n=0^N x(n)* e^{-i*2*pi*n*k/N}
#
# => the Convolution theorem then writes :
#    ##########################################################
#    ##                                                    ####
#    ##     DFT(x conv y) = N * X*Y = N * DFT(x)*DFT(y)    ####
#    ##                                                    ####
#    ##     DFT(x * y) = X conv Y = DFT(x) conv DFT(y)     ####
#    ##                                                    ####
#    ##########################################################
#

# ---- Spectrum of B = spectrum of enveloppe ---------------
Spec_env = np.abs(np.fft.fft(B,norm='forward'))**2
Spec_env_shifted = np.fft.fftshift(Spec_env)

# ---- compute DFT for eta_c, eta_c* and eta_r -------------
DFT_eta_c0 = np.fft.fft(eta_c,norm='forward')
DFT_eta_c_conj0 = np.fft.fft(np.conjugate(eta_c),norm='forward')
DFT_eta_c = np.fft.fftshift(DFT_eta_c0)
DFT_eta_c_conj = np.fft.fftshift(DFT_eta_c_conj0)

DFT_eta_r0 = np.abs(np.fft.fft(eta_r,norm='forward'))
DFT_eta_r = np.fft.fftshift(DFT_eta_r0)
DFT_eta_rold0 = (np.fft.fft(eta_r,norm='forward'))
DFT_eta_rold = np.fft.fftshift(DFT_eta_rold0)


### Convolve directly the DFT of eta_c and eta_c* = eta_c_conj ####################
#    DFT(|A|**2) = DFT(eta_c * eta_c_conj) = DFT(eta_c) conv DFT(eta_c_conj)
conv_DFT0 = np.convolve(DFT_eta_c,DFT_eta_c_conj,'full')
conv_DFT = np.abs(conv_DFT0[n//2:n//2+n])**2
conv_DFT[n//2]=Spec_env_shifted[n//2]

### Use the identity : DFT(eta_c) = 2 * DFT(eta)_+, 
#  where DFT(.)_+ is the DFT(.) with values =0 for negative frequencies 
#  and DFT(.)_- is the DFT(.) with values =0 for positive frequencies
#
# /!\ watch out !!!! when initializing a vector you must specify dtype=np.complex128
# if you don't specify that it is meant to be a complex vector the im part is purely removed !!!!
#
DFT_eta_r_pos = np.zeros(DFT_eta_r.shape,dtype=np.complex128)
DFT_eta_r_pos[n//2:] = DFT_eta_rold[n//2:]#*(1+1j)*np.sqrt(2)

DFT_eta_r_neg = np.zeros(DFT_eta_r.shape,dtype=np.complex128)
DFT_eta_r_neg[0:n//2] = DFT_eta_rold[0:n//2]#/np.sqrt(2**3)

###  Write the convolution using DFT(eta)_+ and DFT(eta)_-, instead of DFT(eta_c)
conv_DFT_bis0 = np.convolve(2*DFT_eta_r_pos,2*DFT_eta_r_neg,'full')
conv_DFT_bis = conv_DFT_bis0[n//2-1:n//2-1+n]
conv_DFT_bis[n//2]=Spec_env_shifted[n//2]

###  Write the convolution using DFT(eta)_+ ONLY, instead of DFT(eta_c) (using the symmetry of DFT(eta))
conv_DFT_ter0 = np.convolve(2*DFT_eta_r_pos,2*np.flip(DFT_eta_r_pos),'full')
conv_DFT_ter = conv_DFT_ter0[n//2-1:n//2-1+n]

### take the modulus squared
Spec_calculated_env = np.abs(conv_DFT_bis)**2
Spec_calculated_env[n//2]=Spec_env_shifted[n//2]
Spec_calculated_env_pos = np.abs(conv_DFT_ter)**2 
Spec_calculated_env_pos[n//2]=Spec_env_shifted[n//2]

kcycles = freq_shifted[:]/(2*np.pi)

moment_1 = np.nansum(dkx*(kcycles[n//2:])*Spec_env_shifted[n//2:])
moment_0 = np.nansum(dkx*Spec_env_shifted[n//2:])
echelle_env= moment_0/moment_1

moment_1 = np.nansum(dkx*(kcycles[n//2:])*Spec_calculated_env_pos[n//2:])
moment_0 = np.nansum(dkx*Spec_calculated_env_pos[n//2:])
echelle_mod= moment_0/moment_1

print("Echelle du spectre de l'enveloppe : ",echelle_env)
print("Echelle du spectre de l'enveloppe modÃ©lisÃ© : ",echelle_mod)
############################  FIGURES ###################################################################
# --- Figure of the position of the problem: enveloppe spectrum vs convolutions #########################
fig,axs=plt.subplots(1,2,figsize=(20,6), gridspec_kw={'width_ratios': [3, 1]})
axs[0].plot(Tvec,eta_r,label='signal $\eta(t)$')
axs[0].plot(Tvec,np.abs(eta_c),label='enveloppe $|\eta_c(t)|$')
axs[0].grid(True)
axs[0].set_xlabel('t [s]')
axs[0].set_ylabel('$\eta$ [m]')
axs[0].legend()

axs[1].plot(freq_shifted,np.abs(DFT_eta_c),'o',label='Spec $\eta_c$')
axs[1].plot(freq_shifted,np.abs(DFT_eta_c_conj),'o',label='Spec $\eta_c$*')
axs[1].plot(freq_shifted,2*np.abs(DFT_eta_r),'.',label='2*Spec $\eta$')
axs[1].grid(True)
axs[1].set_xlim((-0.2,0.2))
axs[1].legend()

# --- Figure of what we want : enveloppe spectrum vs convolutions #######################################
fig,axs=plt.subplots(1,2,figsize=(20,6),gridspec_kw={'width_ratios': [2, 1]})
_=axs[0].plot(freq_shifted,Spec_env_shifted,'o',label='Spec')
_=axs[0].plot(freq_shifted,conv_DFT,'*',label='Spec via convolution')
_=axs[0].plot(freq_shifted,Spec_calculated_env,'.',label='Spec from DFT($\eta$)$_+$ and DFT($\eta$)$_-$')
_=axs[0].plot(freq_shifted,Spec_calculated_env_pos,'.',label='Spec from DFT($\eta$)$_+$ only')
_=axs[0].grid(True)
# _=plt.ylim((10**-10,10**2))
_=axs[0].set_xlabel('k')
_=axs[0].set_yscale('log')
_=axs[0].legend()

# --- Figure of ratio between spec env / spec calculated ##############################################
_=axs[1].plot(freq_shifted,Spec_env_shifted/Spec_calculated_env,'.',label='ratio env/calculated')
_=axs[1].plot(freq_shifted,Spec_env_shifted/Spec_calculated_env_pos,'.',label='ratio env/calculated pos only')
_=axs[1].plot([freq_shifted[n//4],freq_shifted[n//4]],[10**-2,10**18],'--r')
_=axs[1].plot([freq_shifted[3*n//4],freq_shifted[3*n//4]],[10**-2,10**18],'--r')
_=axs[1].grid(True)
_=axs[1].set_xlabel('k')
_=axs[1].set_yscale('log')
_=axs[1].legend()

# --- Figure of spectrum(eta) with eta the real signal (positive and negative parts) ###################
_=plt.figure(figsize=(12,6))
_=plt.plot(freq_shifted,np.real(DFT_eta_r_pos),'o',label='Real DFT($\eta$)$_+$')
_=plt.plot(freq_shifted,np.real(DFT_eta_r_neg),'o',label='Real DFT($\eta$)$_-$')
_=plt.plot(freq_shifted,np.imag(DFT_eta_r_pos),'.',label='Imag DFT($\eta$)$_+$')
_=plt.plot(freq_shifted,np.imag(DFT_eta_r_neg),'.',label='Imag DFT($\eta$)$_-$')
_=plt.grid(True)
_=plt.yscale('log')
_=plt.legend()

med = np.median(Spec_env_shifted[n//4:3*n//4]/Spec_calculated_env_pos[n//4:3*n//4])
print('med = ',med)
```

```python
#### Investigate "std of surface" #########################
nsee=1000
nH = 20
nsig = 10

pt_min = 6

####### Set-up the problem ####################################""
dt = 1
n = 2048
Tvec=dt*np.arange(0,n)
dk = 2*np.pi/(dt*n)

# sig = 0.001
f = np.fft.fftfreq(n, d=dt)
freq_shifted = np.fft.fftshift(f)
df = freq_shifted[1]-freq_shifted[0]
Tp = 16
fp = 1/Tp
kp = (2*np.pi)**2*fp**2/9.81
Hss = np.linspace(1,20,nH)
sigs= np.logspace(-2,np.log(0.2),nsig)
ntile_max = n//pt_min
tiles = np.arange(2,ntile_max+1)
ntile = len(tiles)
print('ntile = ',ntile)
Hs_surf_std = np.zeros((nH,nsig,ntile,nsee))
Hs_surf_mean = np.zeros((nH,nsig,ntile,nsee))
Hs_spec = np.zeros((nH,nsig,nsee))
Hs_env = np.zeros((nH,nsig,nsee))
mean_env = np.zeros((nH,nsig,nsee))

for see in range(1):#nsee):
    rng = np.random.default_rng(see)
    rg = rng.uniform(low=0.0, high=1.0, size=(n))
    for ih, Hs in enumerate(Hss):
        for isig,sig0 in enumerate(sigs):
            sig = sig0*fp
            ##Â Define a spectrum #################################
            E_k = (Hs/4)**2/(sig*np.sqrt(2*np.pi))*np.exp((-0.5*(f-fp)**2/sig**2))
            zhats=(np.sqrt(2*E_k*df)*np.exp(1j*2*np.pi*rg))
            eta_c = np.fft.fftshift(np.fft.ifft(zhats,norm="forward"))
            eta_r = np.real(eta_c)
            eta_i = np.imag(eta_c)

            env0 = np.sqrt(eta_r**2+eta_i**2)
            env = env0 - env0.mean()
            for iti, numtiles in enumerate(tiles):
                Hs_surf0 = np.zeros(numtiles)
                sht = int(n//numtiles)
                for ktile in range(numtiles):
                    Hs_surf0[ktile] = 4 * np.std(eta_r[int(ktile*sht):int((ktile+1)*sht)])
                                    
                Hs_surf_std[ih,isig,iti,see] = np.std(Hs_surf0)
                Hs_surf_mean[ih,isig,iti,see] = np.mean(Hs_surf0)
            Hs_env[ih,isig,see] = 4 * np.std(env)
            mean_env[ih,isig,see] = np.mean(env0)
            Hs_spec[ih,isig,see] = 4 * np.sqrt(sum(E_k.flatten()*df)) 
```

```python
#####  Investigate variations with different seed numbers #########################""
ns = 60

sum_abs = np.zeros(ns)
sum_re = np.zeros(ns)
sum_im = np.zeros(ns)

abs_DFT = np.zeros((n,ns))

for see in range(ns):
    rng = np.random.default_rng(see)
    rg = rng.uniform(low=0.0, high=1.0, size=(n))

    zhats=(np.sqrt(2*E_k*dk)*np.exp(1j*2*np.pi*rg))
    eta_c = np.fft.fftshift(np.fft.ifft(zhats,norm="forward"))
    eta_r = np.real(eta_c)
    eta_i = np.imag(eta_c)
    DFT_eta_r0 = (np.fft.fft(eta_r,norm='forward'))
    DFT_eta_rold = np.fft.fftshift(DFT_eta_r0)

    sum_abs[see] = np.sum(np.abs(DFT_eta_rold))
    sum_re[see] = np.sum(np.real(DFT_eta_rold))
    sum_im[see] = np.sum(np.imag(DFT_eta_rold))
    abs_DFT[:,see] = np.abs(DFT_eta_rold)

fig,axs=plt.subplots(1,3,figsize=((20,6)))
axs[0].plot(np.arange(ns),sum_abs/np.mean(sum_abs),label='sum |DFT|')
axs[0].grid(True)
axs[0].set_xlabel('seed number')
# axs[0].set_title('sum |DFT|')
axs[1].plot(np.arange(ns),sum_re/np.mean(sum_re),label='sum re(DFT)')
axs[1].grid(True)
axs[1].set_xlabel('seed number')
axs[1].set_title('sum re(DFT)')
axs[2].plot(np.arange(ns),sum_im/np.mean(sum_im),label='sum im(DFT)')
axs[2].grid(True)
axs[2].set_xlabel('seed number')
axs[2].set_title('sum im(DFT)')

mean_abs_DFT = np.mean(abs_DFT,axis=1)

plt.figure(figsize=(16,6))
colors = plt.cm.jet(np.linspace(0,1,ns))
for see in range(ns):
    plt.plot(freq_shifted,abs_DFT[:,see],color=colors[see],linewidth=2)

plt.plot(freq_shifted,mean_abs_DFT,'--w')
plt.grid(True)
plt.yscale('log')
```
