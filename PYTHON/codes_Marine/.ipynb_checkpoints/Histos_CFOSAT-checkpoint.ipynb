```python
savename = PATH_storage+'Hs_vs_std_boxes_histo_CFOSAT.npz'
data = np.load(savename)

for n in data.keys():
    print(n+'_boxes = data["'+n+'"]')
    exec(n+'_boxes = data["'+n+'"]')

im=plt.pcolormesh(Hs_bins_boxes[:-1], Hs_std_bins_boxes[:-1], np.log(Hs_vs_std_histo_boxes.T))

_=plt.colorbar(im)
_=plt.xlabel('Hs')
_=plt.ylabel('std Hs')
_=plt.title('boxes')

Hs_2D,Hs_std_2D = np.meshgrid(Hs_bins_boxes[1:-1],Hs_std_bins_boxes[1:-1])
ratio_Hs_vs_std_2D = Hs_2D/Hs_std_2D
ratio_Hs_vs_std_2D[Hs_std_2D==0]=0
Hs_vs_std_histo_boxes2 = Hs_vs_std_histo_boxes.T
r_min = ratio_Hs_vs_std_2D[ratio_Hs_vs_std_2D>0].min()
r_max = ratio_Hs_vs_std_2D.max()
# print('r_min = ',r_min,', r_max = ',r_max)
ratio_bins = np.logspace(np.log10(10**-2),np.log10(303),101)
# print('ratio_bin min = ',ratio_bins.min(),', ratio_bin max = ',ratio_bins.max())

def fmt(x):
#     s = f"{np.exp(x):.0f}"
    if np.exp(x)<1:
        s=f"{np.exp(x):.3f}"
    else:
        s = f"{np.exp(x):.0f}"
    return rf"{s}"#" \%" if plt.rcParams["text.usetex"] else f"{s} %"

CS=plt.contour(Hs_2D,Hs_std_2D,np.log(ratio_Hs_vs_std_2D),levels=10)
_=CS.clabel(fmt=fmt)
_=plt.grid(True)

Hs_over_std_Histo = np.zeros(len(ratio_bins)-1)
for ir,r in enumerate(ratio_bins[:-1]):
    ind=np.where((ratio_Hs_vs_std_2D[:-1,:-1]>=r)& (ratio_Hs_vs_std_2D[:-1,:-1]<ratio_bins[ir+1]))
#     print(np.shape(ind[0]))
#     print('ind[0] = ',ind[0],', ind[1] = ',ind[1])
#     print(ratio_Hs_vs_std_2D.shape)
#     print('ir = ',ir)
    Hs_over_std_Histo[ir] = np.sum(Hs_vs_std_histo_boxes2[ind[0],ind[1]])
# _=plt.yscale('log')

plt.figure()
plt.hist(ratio_bins[:-2], ratio_bins[:-1], weights=Hs_over_std_Histo[:-1])
_=plt.yscale('log')
ind = np.where(Hs_over_std_Histo[:-1]>0)[0]
ratio_lim = ratio_bins[ind[-1]+1]
_=plt.axvline(ratio_lim,color='r')
print('value ind = ',ind[-1],', ratio = ',ratio_bins[ind[-1]+1])

print("--- ",data['num_tracks_good'], " tracks looked at / ",data['num_tracks']," total tracks ------------------- ")
```
